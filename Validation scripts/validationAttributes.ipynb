```python
dfAttributes=pd.read_csv('gdrive/My Drive/NIER-ICSE/data/test/attributes.csv')
dfAttributes['Results2_partialDiagram']=dfAttributes['Results2_partialDiagram'].apply(lambda x: x.split(','))
dfAttributes['attributes']=dfAttributes['attributes'].apply(lambda x: eval(x))
dfAttributes.fillna("",inplace=True)
dfAttributes['manualResults']=dfAttributes['manualResults'].apply(lambda x: x.split(','))
#dfAttributes['manualResults'][155]=[' name',' surname', ' nextDestination', ' phoneNumber', ' nextDestination', 'email']
```

```python
def removeEmptyString(listString):
  for idx,i in enumerate(listString): 
    if i=="": 
      listString.remove(i)
  return listString
dfAttributes['manualResults']=dfAttributes['manualResults'].apply(lambda x: removeEmptyString(x))
```

```python
def Recall(predictedList,GroundTruthList): #TP+/FN+TP 
  score=0  
  
  GroundTruthList=list(map(str.lower,GroundTruthList))
  for i in predictedList: 
    i=i.strip()
    if i.lower() in GroundTruthList : 
     
      score=score+1
  x=len(GroundTruthList)
  if x==0 : 
      return 0
  return score/len(GroundTruthList)
```

```python
dfAttributes['Exact_recall']=dfAttributes.apply(lambda x: Recall(x['Results2_partialDiagram'],(x['attributes']) ),axis=1)
#dfAttributes['precision']=dfAttributes.apply(lambda x: precision(x['Results2_partialDiagram'],(x['attributes']) ),axis=1)
```

```python
dfAttributes['Similar_recall']=dfAttributes.apply(lambda x: len((x['manualResults']) )/len(x['attributes']),axis=1)
dfAttributes
```
