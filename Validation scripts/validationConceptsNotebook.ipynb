```python
import pandas as pd
import re
import sys
import os
import openai
import wordninja



df=pd.read_csv('gdrive/My Drive/NIER-ICSE/data/test/seqConcepts.csv')
data=''
for i, row in df.iterrows():
  data= data+ '\n' + str(row['sequence']).strip('{}')+ '.'



def interceptList(EdConcepts):
    allCouples = [(a, b) for idx, a in enumerate(EdConcepts) for b in EdConcepts[idx + 1:]]

    return allCouples


def removeLetter(TheList):
    for ind, j in enumerate(TheList):

        if len(j) == 1:
            TheList.remove(j)
    return TheList


def removeDuplicated(TheList):
    for ind, j in enumerate(TheList):
        for indx, i in enumerate(TheList[ind + 1:]):
            if i.lower() == j.lower():
                TheList.remove(j)
    return TheList


def retrieveConcepts(res):
    S = re.sub(r'\W+', ' ', res)
    FirstSet = S.split(' ')
    NewConcepts = wordninja.split(S)
    NewConcepts.extend(FirstSet)
    NewConcepts = ' '.join(NewConcepts).split()
    concepts = set(NewConcepts)
    concepts = list(concepts)
    concepts = removeLetter(concepts)
    concepts = removeDuplicated(concepts)

    return concepts

def predictFinalList(designList_):
    Prompt = 'Continue the line: \n ' + data + '\n' + str(designList_) + ','

    response_ = openai.Completion.create(
        engine="text-davinci-002",
        prompt=Prompt,
        temperature=0.7,
        max_tokens=15,
        top_p=1,
        frequency_penalty=0,
        presence_penalty=0
    )

    res_ = response_.choices[0].text
    concepts = retrieveConcepts(str(designList_) + ',' + res_)

    return concepts, res_



```

##Metrics

###Exact match

######  recall

```python
def Recall(predictedList,GroundTruthList): #TP+/FN+TP 
  score=0  
  
  GroundTruthList=list(map(str.lower,GroundTruthList))
  for i in predictedList: 
    
    if i.lower() in GroundTruthList : 
     
      score=score+1
  x=len(GroundTruthList)
  if x==0 : 
      return 0
  return score/len(GroundTruthList)
```

#### precision

```python

def precision (predictedList,testList): #TP+/FP+TP 
  score=0 

  testList=list(map(str.lower,testList))
  
  for i in predictedList: 
    i=i.lower()
    if i in testList : 
     
      score=score+1
  x=len(predictedList)
  if x==0 : 
      return 0
  return score/len(predictedList)
```

#### Similarity match 

```python
def findSynonyms(word):
  
  synonyms = []
  for syn in wordnet.synsets(word.lower()):
      for l in syn.lemmas():
        synonyms.append(l.name())
  return list(set(synonyms))
  

```

```python
def similarMatch(predictedList, labeldList):

  score=0
  for i in predictedList: 
    
    for j in labeldList: 
     
      synonyms=findSynonyms(j)
      if i.lower() in synonyms : 
        score=score+1
        
  return score/len(labeldList)
```

```python
import re


def parseResults(results):
  suggestions= results.split('\n')
  concepts=[]
  for s in suggestions: 
    concepts.append(s.split('=')[0])
  return concepts
```

```python
def createHashtag(results): 
  
  concepts={}
  suggestions= results.split('\n')
  for s in suggestions: 
    key = s.split('=')[0]
    value = s.split('=')[1]
    concepts[key]=value
  return concepts
```

```python
df['maps']=df['results'].apply(lambda x: createHashtag(x))
```

```python
df['suggestions']=df['results'].apply(lambda x: parseResults(x) )
```

```python
def selectedConcepts(conceptsMap, attempts): 
  new= dict(sorted(conceptsMap.items(), key=lambda item: item[1],reverse=True))

  selected=[]
  for idx, x in enumerate(list(new)): 
    if (idx<attempts): 
      selected.append(x)
    else : 
      break 
  
  return selected
```

```python
df['selectedConcepts']=df.apply(lambda x: selectedConcepts(x['maps'],x['attempts'] ),axis=1)
```
